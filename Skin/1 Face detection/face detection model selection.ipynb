# Haar Cascade

    장점
        CPU에서 거의 실시간으로 작동
        간단한 아키텍처
        다른 스케일로 얼굴을 감지
        
    단점
        이 방법의 주요 단점은 False 예측이 많다는 것입니다.
        정면이 아닌 이미지에서는 작동하지 않습니다.
        폐색에서 작동하지 않습니다


# DNN
    
    장점
        네 가지 방법 중 가장 정확
        CPU에서 실시간으로 실행
        위, 아래, 왼쪽, 오른쪽, 측면 등 다양한 방향으로 작동
        상당한 폐색에서도 작동
        다양한 스케일에서 얼굴을 감지(작은 얼굴뿐만 아니라 큰 얼굴도 감지)
        
    DNN 기반 검출기는 Haar가 제공하는 이점을 그대로 유지하면서 Haar 캐스케이드 기반 검출기의 모든 단점을 커버 가능. 
    이 방법이 다음에 논의 할 Dlib HoG 기반 얼굴 탐지기보다 느리다는 것을 제외하고는이 방법의 큰 단점을 찾기 어려움
    

# Dlib HoG

    장점
        CPU에서 가장 빠른 방법
        정면 및 약간 정면이 아닌 얼굴에 매우 효과적
        다른 세 가지에 비해 가벼운 모델.
        작은 폐색에서 작동
        
    단점
        가장 큰 단점은 최소 얼굴 크기 80 × 80으로 훈련 되었기 때문에 작은 얼굴을 감지하지 못함. 
        따라서 응용 프로그램의 얼굴 크기보다 얼굴 크기가 커야한다. 
        그러나 더 작은 크기의 얼굴에 대해 자신 만의 얼굴 탐지기를 훈련시킬 수 있다.
        경계 상자는 종종 이마의 일부와 때로는 턱의 일부도 제외됨.
        상당한 폐색에서 잘 작동하지 않는다
        아래 또는 위와 같이 측면과 정면이 아닌 얼굴에는 작동하지 않는다.


# Dlib CNN

    장점
        다른 얼굴 방향으로 작동
        폐색에 견고 함
        GPU에서 매우 빠르게 작동
        매우 쉬운 교육 과정
        
    단점
        CPU에서 매우 느림
        최소 얼굴 크기가 80 × 80 인 훈련 된 작은 얼굴은 감지하지 않음. 
        따라서 응용 프로그램의 얼굴 크기보다 얼굴 크기가 커야한다. 
        그러나 더 작은 크기의 얼굴에 대해 자신 만의 얼굴 탐지기를 훈련시킬 수 있다.
        바운딩 박스는 HoG 검출기보다 훨씬 작다.
