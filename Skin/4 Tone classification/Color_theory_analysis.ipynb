# 색분석 정리
   
   
   
## 1. 색이론

- Color System
    - `색공간`(≒색체계) : 색상환(사람이 인식하는 색들을 비슷한 순서로 붙여 원을 이룬 배열)의 중심축에서 수평방향으로 채도 변화를 표현하고, 수직방향으로 명도 변화를 표현한 원통형태의 3차원 색좌표   
    - `Munsell color system` : 최초로 고안된 색공간. 색상(Hue), 명도(Value), 채도(Chroma) 속성으로 표시   
    - `RGB` : 삼원색(Red, Green, Blue)를 합치면 흰색이 되는 빛의 특성을 이용해 색을 표현하는 가산혼합법   
    - `HSV` : 인간의 시각 특징을 고려해 고안된 색상(Hue), 채도(Saturation), 명도(Value)로 색을 표현하는 방법   
- 색채 대비 이론 : 특정 색 요소의 특징을 강조하기 위해 그와 상반되는 형태, 색채, 톤을 나란히 배치하는 행위   
    - `색상대비` : 서로 색상이 다른 색을 비교하여 발생하는 차이로, 보색과 배치하였을 때 가장 강조됩니다.   
    - `명도대비` : 색의 명암으로 인해 발생하는 차이로, 밝은색과 어두운 색을 함께 배치하면 밝은 색은 더 밝게, 어두움색은 더 어둡게 강조됩니다.   
    - `채도대비` : 색의 채도로 인해 발생하는 차이로, 채도가 높은 색은 더 높게 낮은 색은 더 낮게 강조됩니다.
- 색 조화 이론   
    - `보색대비` : 색상환의 반대편   
    - `근접보색대비` : 색상환의 반대편인 보색을 기준으로 양 쪽으로 색상   
    - `삼각보색대비` : 색상환에서 삼각형을 형성하면서 서로 등거리에 있는 3가지 색상
    - `이중보색대비` : 기준색으로부터 선을 그으면 직사각형 모양을 띄는 4색대비   

[Munsell Color Palette](http://pteromys.melonisland.net/munsell/) 먼셀표기법에 따른 색상 확인 가능   
[Adobe Color](https://color.adobe.com/ko/create) RGB, HSV 등 표기법에 따른 조화색상 확인 가능   
   
   
   
## 2. 퍼스널컬러 이론

- `퍼스널컬러`는 개인 고유의 신체색을 의미하며, 개인의 다양한 특성을 분석하여 퍼스널컬러 타입 분류체계의 한 타입에 개인을 대응시켜 그에 조화를 이루는 색을 진단하여 각 개인에게 적합한 메이크업, 헤어, 의상 등의 색채 이미지를 연출하는 시스템입니다.   
- 기초적으로 색채대비와 착시현상의 이론적인 내용에 따르며, 밝고 생기있는 긍정적인 이미지를 형성하는 데 도움을 주는 것을 목적으로 합니다. 하지만 이러한 퍼스널컬러 시스템은 개인의 특징을 진단하는 기준이 뚜렷하지 않고, 평가 방법이 관능평가(사람이 감각기관을 통해 조사하는 것)인 한계점을 가지고 있습니다.   
- 퍼스널컬러 타입 분류 체계   
현재 한국에서 통용되는 퍼스널 컬러 분류 기준은 '동양인 대상 분류 체계안'과 'PCCS 색 체계'를 기반으로 한 분류 기준   
※ '동양인 대상 분류 체계안' : Fujii and von Alten(1992)에서 제시된 기준   
※ 'PCCS 색 체계' : 일본 색채 연구소가 1964년 발표한 색채 조화를 목적으로 한 컬러 시스템   
    - `Warm Spring bright` : 순색의 Yellow base 톤을 포함한 고채도의 선명한 색   
    - `Warm Spring light` : 흰색이 포함된 Yellow base 톤의 저채도와 고명도의 색   
    - `Warm Autumn mute` : 회색이 포함된 Yellow base 톤의 중-저채도와 중-저명도의 색   
    - `Warm Autumn deep` : 검은색이 섞인 Yellow base 톤의 고-저채도와 중-저명도의 색   
    - `Cool Summer light` : 흰색이 포함된 Blue base 톤의 저채도와 고명도의 색   
    - `Cool Summer mute` : 회색을 포함하는 Blue base톤의 저채도와 중-저명도의 색   
    - `Cool Winter deep` : 검은색이 섞인 Blue base 톤의 고-저채도와 저명도의 색   
    - `Cool Winter bright` : 순색의 Blue base 톤을 포함한 고채도의 선명한 색   
   
   
   
## 3. 분석내용
### 1) 선행연구
> Fujii and von Alten(1992) '퍼스널컬러 유형 분류의 독립변수는 안면 색'   
> '여성들이 이마, 턱에 비해 볼의 황색도를 보정하지 않는다' → '여성이 볼의 톤이 자신의 톤을 대표한다고 판단'   
> 따라서 퍼스널 컬러 유형 분류 시스템의 독립변수로 안면부의 볼의 스킨 톤을 특정   
> 김용현·오유석·이정훈. 퍼스널 컬러 스킨 톤 유형 분류의 정량적 평가 모델 구축에 대한 연구. 한국의류학회지, 42(1), 121~132   
    
> 문완묵(2005)의 연구 '한국인의 피부색 특징은 2.5YR-1.5Y, 명도 6.4-8.5, 채도는 3 내외로 저채도에 속한다'   
> 신향선(2016)의 연구 '색상과 명도 기준으로 6.5YR-7.5YR 계열 중 명도 7.5 이상을 봄, 7.0이하를 가을, 2.5YR-5.5YR 계열 중에서 7.0이상을 여름 3.5YR-5.5YR 계열의 6.5이하는 겨울'   
> '봄'과 '여름' 피부색을 고명도, '가을'과 '겨울'의 피부색을 저명도로 구분(김용현 외, 2018; 정연아, 2016; 신향선, 2016)   
> 이소영(2019), 퍼스널컬러의 톤 유형과 피부색에 따른 조화색의 범주화. 홍익대학교 문화정보정책대학원 석사학위논문.   
   
- 선행연구에서 공통된 타입 분류 기준   
    - 1차 분류기준 : 색(Hue) 온도   
        - Yellow base : Warm tone   
        - Blue base : Cool tone   
    - 2차 분류기준 : 명도(Value)   
        - High Value : Spring, Summer   
        - Low Value : Autumn, Winter   
    - 3차 분류기준 : 채도(Saturation)   
        - High Saturation : bright, deep   
        - Low Saturation : light, mute   
        
### 2) 선행 타입분류 기준 문제점
1. MUNSELL Color System   
    색을 표현하는 방법에 문자가 포함됨   
    → 알고리즘을 구현, 딥러닝 적용에 한계가 있음   
    색을 구분하는 단계가 세부적이지 않으며 색마다 표현 가능한 단계가 차이가 있음   
    → 1~3차 기준 모두 명확하게 분류할 수 없음   
    
2. Hue에 따라 변경되는 색의 밝기   
    명도 값이 같아도 색상 값이 다른 경우 색의 밝기가 달라짐   
    따라서 2차 분류기준에서 특정 명도 값을 기준으로 분류할 경우 오차범위가 넓음   
    → 밝기(명암)를 가장 명확하게 구분할 수 있는 수치 필요

### 3) 타입 분류 기준 변경
1. 타입분류에 사용될 색 체계 변경   
    - 알고리즘 구현과 딥러닝 적용이 가능한 색 체계로 변환 필요   
    - RGB, HSV로 변환해서 퍼스널컬러 타입 분류기준과 상관관계 테스트   
        - RGB   
            RGB로 표현할 수 있는 색의 범위가 더 넓어 여러 연산과정을 거칠 경우 HSV보다 더 정확한 결과값을 추출할 수 있음   
            하지만 각 red, green, blue의 정도값을 표현하기 때문에 채도, 명도가 r, g, b의 수치 변경에 영향을 많이 받음   
        - HSV   
            퍼스널 컬러 타입 분류 기준인 3가지 항목을 모두 가지고 있음   
            RGB보다는 색 표현 범위가 작지만 먼셀 색 체계보다 정확한 표현이 가능함   
            
    → HSV로 타입 분류에 사용할 색 체계 선정   
   
   
2. 2차 분류 기준 변경
    - H, S, V 항목 별 상관관계 분석_명암테스트   
    - 명도(Value) - 채도(Saturation)값이 클 수록 색상 값의 정도와 관계 없이 명암 차이가 가장 균일하게 밝아짐   
    ※ 하단 명암테스트 이미지 참고   
    
    → 명도(Value) 기준에서 명암(Value-Saturation) 기준으로 변경
