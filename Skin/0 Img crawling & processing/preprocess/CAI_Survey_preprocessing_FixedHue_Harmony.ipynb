### hueFixed_harmony + class + cnt

```python
## 밝은색 설문 결과 불러오기
hueFixed_bright = pd.read_excel("C:/Users/cho ye seul/Documents/예슬/Acorn/FinalProject/Statistical Analysis/퍼스널컬러 정량화 모델 가설검정을 위한 설문 (2).xlsx",
                          sheet_name = 'Sheet1', header=0,na_values = 'NaN',
                          usecols="M, N, O, P, Q, R")

hueFixed_bright.rename(columns={hueFixed_bright.columns[0]:"comp",
                                hueFixed_bright.columns[1]:"triangle1",
                                hueFixed_bright.columns[2]:"triangle2",
                                hueFixed_bright.columns[3]:"doubleness1",
                                hueFixed_bright.columns[4]:"doubleness2",
                                hueFixed_bright.columns[5]:"doubleness3"}, inplace=True)

columns = ["comp", "triangle1", "triangle2", "doubleness1", "doubleness2", "doubleness3"]

## 선택항목 별 부위 매칭 (①-채도(-5)/명도(-5), ②-채도(-5)/명도(+5), ③-original, ④-채도(+5)/명도(+5), ⑤-채도(+5)/명도(-5))
for i in range(int(hueFixed_bright.shape[0])):
    for col in columns:
        if hueFixed_bright[col][i] == "①":
            hueFixed_bright[col][i] = "-S-V"
        elif hueFixed_bright[col][i] == "②":
            hueFixed_bright[col][i] = "-S+V"
        elif hueFixed_bright[col][i] == "③":
            hueFixed_bright[col][i] = "original"
        elif hueFixed_bright[col][i] == "④":
            hueFixed_bright[col][i] = "+S+V"
        else:
            hueFixed_bright[col][i] = "+S-V"

            
## 하나씩 원하는 df형태로 만들기 
# comp 
c_hFixed  = pd.DataFrame(hueFixed_bright["comp"])
c_hFixed["class"] = "comp"
c_hFixed.rename(columns={"comp":"select"}, inplace=True)

# triangle1
t1_hFixed  = pd.DataFrame(hueFixed_bright["triangle1"])
t1_hFixed["class"] = "triangle1"
t1_hFixed.rename(columns={"triangle1":"select"}, inplace=True)

# triangle2
t2_hFixed  = pd.DataFrame(hueFixed_bright["triangle2"])
t2_hFixed["class"] = "triangle2"
t2_hFixed.rename(columns={"triangle2":"select"}, inplace=True)

# doubleness1
d1_hFixed  = pd.DataFrame(hueFixed_bright["doubleness1"])
d1_hFixed["class"] = "doubleness1"
d1_hFixed.rename(columns={"doubleness1":"select"}, inplace=True)

# doubleness2
d2_hFixed  = pd.DataFrame(hueFixed_bright["doubleness2"])
d2_hFixed["class"] = "doubleness2"
d2_hFixed.rename(columns={"doubleness2":"select"}, inplace=True)

# doubleness3
d3_hFixed  = pd.DataFrame(hueFixed_bright["doubleness3"])
d3_hFixed["class"] = "doubleness3"
d3_hFixed.rename(columns={"doubleness3":"select"}, inplace=True)


## 다 합치기
hFixed_harmony = pd.concat([c_hFixed, t1_hFixed, t2_hFixed, d1_hFixed, d2_hFixed, d3_hFixed])

# 개수 넣기
hFixed_harmony["cnt"] = 1

hFixed_harmony
```

```python
# 기본항목 설문결과 (모집단 분리하지 않은 항목) 데이터 준비

## 밝은색 설문 결과 불러오기
elementaryItem = pd.read_excel("C:/Users/cho ye seul/Documents/예슬/Acorn/FinalProject/Statistical Analysis/퍼스널컬러 정량화 모델 가설검정을 위한 설문 (2).xlsx",
                          sheet_name = 'Sheet1', header=0,na_values = 'NaN',
                          usecols="C, D, M, N, O, P, Q, R")
elementaryItem.rename(columns={elementaryItem.columns[0]:"sex",
                          elementaryItem.columns[1]:"age_class",
                          elementaryItem.columns[2]:"comp",
                          elementaryItem.columns[3]:"triangle1",
                          elementaryItem.columns[4]:"triangle2",
                          elementaryItem.columns[5]:"doubleness1",
                          elementaryItem.columns[6]:"doubleness2",
                          elementaryItem.columns[7]:"doubleness3"}, inplace=True)


columns = ["comp", "triangle1", "triangle2", "doubleness1", "doubleness2", "doubleness3"]
## 선택항목 별 부위 매칭 (①-채도(-5)/명도(-5), ②-채도(-5)/명도(+5), ③-original, ④-채도(+5)/명도(+5), ⑤-채도(+5)/명도(-5))
for i in range(int(elementaryItem.shape[0])):
    for col in columns:
        if elementaryItem[col][i] == "①":
            elementaryItem[col][i] = "-S-V"
        elif elementaryItem[col][i] == "②":
            elementaryItem[col][i] = "-S+V"
        elif elementaryItem[col][i] == "③":
            elementaryItem[col][i] = "original"
        elif elementaryItem[col][i] == "④":
            elementaryItem[col][i] = "+S+V"
        else:
            elementaryItem[col][i] = "+S-V"

            
## 값 변경 반복문
for i in range(int(elementaryItem.shape[0])):

    ## sex : Male, Female 리스트 담기
    if elementaryItem["sex"][i] == "남성":
        elementaryItem["sex"][i] = "Male"
    else:
        elementaryItem["sex"][i] = "Female"
        
    ## age_class : 10, 20, 30, 40, 50으로 리스트 담기
    if elementaryItem["age_class"][i] == "10대":
        elementaryItem["age_class"][i] = "10"
    elif elementaryItem["age_class"][i] == "20대":
        elementaryItem["age_class"][i] = "20"
    elif elementaryItem["age_class"][i] == "30대":
        elementaryItem["age_class"][i] = "30"
    elif elementaryItem["age_class"][i] == "40대":
        elementaryItem["age_class"][i] = "40"
    else:
        elementaryItem["age_class"][i] = "50"
       
        
## 하나씩 원하는 df형태로 만들기 
# comp 
c_hFixed  = pd.DataFrame(data = (elementaryItem["sex"], elementaryItem["age_class"], elementaryItem["comp"]))
c_hFixed = c_hFixed.T
c_hFixed.rename(columns={"comp":"select"}, inplace=True)

# triangle1
t1_hFixed  = pd.DataFrame(data = (elementaryItem["sex"], elementaryItem["age_class"], elementaryItem["triangle1"]))
t1_hFixed = t1_hFixed.T
t1_hFixed.rename(columns={"triangle1":"select"}, inplace=True)

# triangle2
t2_hFixed  = pd.DataFrame(data = (elementaryItem["sex"], elementaryItem["age_class"], elementaryItem["triangle2"]))
t2_hFixed = t2_hFixed.T
t2_hFixed.rename(columns={"triangle2":"select"}, inplace=True)

# doubleness1
d1_hFixed  = pd.DataFrame(data = (elementaryItem["sex"], elementaryItem["age_class"], elementaryItem["doubleness1"]))
d1_hFixed = d1_hFixed.T
d1_hFixed.rename(columns={"doubleness1":"select"}, inplace=True)

# doubleness2
d2_hFixed  = pd.DataFrame(data = (elementaryItem["sex"], elementaryItem["age_class"], elementaryItem["doubleness2"]))
d2_hFixed = d2_hFixed.T
d2_hFixed.rename(columns={"doubleness2":"select"}, inplace=True)

# doubleness3
d3_hFixed  = pd.DataFrame(data = (elementaryItem["sex"], elementaryItem["age_class"], elementaryItem["doubleness3"]))
d3_hFixed = d3_hFixed.T
d3_hFixed.rename(columns={"doubleness3":"select"}, inplace=True)


## 다 합치기
SA_hFH_cnt = pd.concat([c_hFixed, t1_hFixed, t2_hFixed, d1_hFixed, d2_hFixed, d3_hFixed])

# 개수 넣기
SA_hFH_cnt["cnt"] = 1

SA_hFH_cnt
```
