### Fashion Interest + Purpose + cnt

```python
## 밝은색 설문 결과 불러오기
elementaryItem = pd.read_excel("C:/Users/cho ye seul/Documents/예슬/Acorn/FinalProject/Statistical Analysis/퍼스널컬러 정량화 모델 가설검정을 위한 설문 (1).xlsx",
                          sheet_name = 'Sheet1', header=0,na_values = 'NaN',
                          usecols="G, H")

elementaryItem.rename(columns={elementaryItem.columns[0]:"purpose",
                               elementaryItem.columns[1]:"interest"}, inplace=True)


## 값 변경 반복문
for i in range(int(elementaryItem.shape[0])):
    ## purpose : clothes&cosmetic(0), clothes(1), cosmetic(2), etc(3)로 리스트 담기
    if ("옷" in elementaryItem["purpose"][i]) & ("화장" in elementaryItem["purpose"][i]):
        elementaryItem["purpose"][i] = "clothes&cosmetic"
    elif ("옷" in elementaryItem["purpose"][i]):
        elementaryItem["purpose"][i] = "clothes"
    elif ("화장" in elementaryItem["purpose"][i]):
        elementaryItem["purpose"][i] = "cosmetic"
    else:
        elementaryItem["purpose"][i] = "etc"

        
# 개수 넣기
elementaryItem["cnt"] = 1

PI_cnt = elementaryItem
PI_cnt
```
