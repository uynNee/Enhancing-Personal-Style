```python
#html 파싱
def get_html(url):
    html = ""
    resp = requests.get(url)
    
    if resp.status_code == 200 :
        html = resp.text
        
    return html
```

```python
#특수문자, 숫자 제거 정규표현식
def cleanText(readData):
 
    #텍스트에 포함되어 있는 특수 문자 제거
    p = re.compile("[^0-9]")
    numData = "".join(p.findall(readData))
    text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', numData)
 
    return text


def crawl_prdt_info() :    
    url = "https://store.musinsa.com/app/product/detail/577636/0"

    #파이어폭스 Headless , 스크립트 처리
    binary = '/usr/bin/firefox'
    options = webdriver.FirefoxOptions()
#     prefs = {'profile.default_content_setting_values': {'cookies' : 2, 
#                                                         'images': 2, 
#                                                         'plugins' : 2, 
#                                                         'popups': 2, 
#                                                         'geolocation': 2, 
#                                                         'notifications' : 2, 
#                                                         'auto_select_certificate': 2, 
#                                                         'fullscreen' : 2, 
#                                                         'mouselock' : 2, 
#                                                         'mixed_script': 2, 
#                                                         'media_stream' : 2, 
#                                                         'media_stream_mic' : 2, 
#                                                         'media_stream_camera': 2, 
#                                                         'protocol_handlers' : 2, 
#                                                         'ppapi_broker' : 2, 
#                                                         'automatic_downloads': 2, 
#                                                         'midi_sysex' : 2, 
#                                                         'push_messaging' : 2, 
#                                                         'ssl_cert_decisions': 2, 
#                                                         'metro_switch_to_desktop' : 2, 
#                                                         'protected_media_identifier': 2, 
#                                                         'app_banner': 2, 
#                                                         'site_engagement' : 2, 
#                                                         'durable_storage' : 2}} 
#     options.add_experimental_option('prefs', prefs)
    
    options.binary = binary
    options.add_argument('start-maximized')
    options.add_argument('--headless')
    options.add_argument("disable-infobars") 
    options.add_argument("--disable-extensions")
    options.headless = True
    
    driver = webdriver.Firefox(options=options, executable_path='/usr/local/bin/geckodriver')
    driver.implicitly_wait(0.1)
    driver.get(url)

    
    c = driver.page_source
    soup = BeautifulSoup(c, "lxml")

    

    ##상품 정보 크롤링
    for div in soup('div', class_="product_article") :
        for p in div("p", class_="product_article_contents") :
            print(p.get_text())

    print("--------------그래프---------------")
    
    #그래프 정보 (조회 현황)

    
    print("[조회현황]")
    for graph in soup("div", class_="graph_area") :
        #bar형 (연령별)
        for ul in graph('ul', class_="bar_wrap") :
            for li in ul('li') :
                print("연령 : " + li.dt.get_text().strip())
                print("비율 : " + li.dd.get_text().replace("\n", "").strip())
        print("--------------------")    
        #원형 (성별)
        for doughnut in graph("div", class_="graph_doughnut_wrap") :
            for p in doughnut("p", class_="graph_doughnut_total_num") :
                print("총 조회수 : " + p.get_text())
            for dl in doughnut("dl", class_="label_info") :
                print()
                print("성별 : " + dl.dt.get_text())
                print("비율 : " + dl.dd.get_text())

                
    #그래프 정보  (구매현황)                
    driver.find_element_by_xpath("//li[@class='graph_title_purchase']").click()
    c = driver.page_source
    soup = BeautifulSoup(c)
    

    print("[구매현황]")
    for graph in soup("div", class_="graph_area") :
        #bar형 (연령별)
        for ul in graph('ul', class_="bar_wrap") :
            for li in ul('li') :
                print("연령 : " + li.dt.get_text().strip())
                print("비율 : " + li.dd.get_text().replace("\n", "").strip())
        print("--------------------")    
        #원형 (성별)
        for doughnut in graph("div", class_="graph_doughnut_wrap") :
            for p in doughnut("p", class_="graph_doughnut_total_num") :
                print("총 구매수 : " + p.get_text())                
            for dl in doughnut("dl", class_="label_info") :
                print("성별 : " + dl.dt.get_text())
                print("비율 : " + dl.dd.get_text())
                
    print("-------------------------------------------")    
```

```python
#view 생성
def create_view() :
    cursor = conn.cursor()
    cursor.execute("DROP VIEW IF EXISTS V_RANK_ALL_LINK;")
    cursor.execute("CREATE VIEW V_RANK_ALL_LINK AS SELECT DISTINCT(PRDT_LINK) AS PRDT_LINK, PRDT_CODE FROM RANK_ALL WHERE PRDT_RANK <= 1000;")
    cursor.execute("SELECT DISTINCT(PRDT_LINK) FROM V_RANK_ALL_LINK;")
    links = []
    for i in cursor :
        links.extend(i)
    cursor.close()
    return links
```
