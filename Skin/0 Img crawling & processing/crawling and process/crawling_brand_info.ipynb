```python
#brand_info
from bs4 import BeautifulSoup
import requests
import pymysql
import re

#특수문자, 숫자 제거 정규표현식
def cleanText(readData):
 
    #텍스트에 포함되어 있는 특수 문자 제거
    p = re.compile("[^0-9]")
    numData = "".join(p.findall(readData))
    text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', numData)
 
    return text
 



def get_html(url):
    html = ""
    resp = requests.get(url)
    
    if resp.status_code == 200 :
        html = resp.text
        
    return html

#좋아요 크롤링
def crawl_musinsa_brand_info(url) :
    brand_name = []
    brand_tab = []
    tab_list = ["상품","룩북","기사","스타일","콘텐츠","후기"]    
    brand_like = []

    link = url
    
    c = get_html(url)
    soup = BeautifulSoup(c)

    
    #브랜드 이름 가져오기
    temp = []
    for brand_div in soup('div', class_="brand_logo") :
        temp.extend(brand_div.p.img['alt'].replace(")", "").split("("))
        brand_name.append(["brd_name_kor" ,temp[0]])
        brand_name.append(["brd_name_eng" ,temp[1]])
        
    dict_name = dict(brand_name)
    
    #상품 탭 내용 크롤링
    for brand_div in soup('div', class_="brand_logo_contents") :
        for brand_li in brand_div('li', class_="box-tab-btn") :
            if( 'class' in brand_li.attrs) :
                  if(cleanText(brand_li.a.get_text()) in tab_list) :     #우리가 뽑은 탭목록에 있을때만 저장
                        brand_tab.append(brand_li.a.get_text().replace(")", "").split("("))   
    
    
    dict_tab = dict(brand_tab)

    
    #좋아요 가져오기
    for brand_div in soup('div', class_="brand_logo_contents") :
        for brand_div2 in brand_div('div', class_="icon_interest") :
            brand_like.append(["like",str(brand_div2.get_text().replace("좋아요", "").strip())])
    
    dict_like = dict(brand_like)

#     print(dict_name)
#     print(dict_tab)
#     print(dict_like)    
  
     #값 테스트
#     print(dict_name["brd_name_eng"], dict_name["brd_name_kor"], 
#           dict_tab.get("상품", 0), dict_tab.get("룩북", 0), dict_tab.get("기사", 0), 
#           dict_tab.get("스타일", 0), dict_tab.get("콘텐츠", 0), dict_tab.get("후기", 0), dict_like.get("like", 0))
    
    #DB 적재
    cursor = conn.cursor()
    data = [dict_name.get("brd_name_eng", str(link.replace("https://store.musinsa.com/app/brand/goods_list/", ""))), 
            dict_name.get("brd_name_kor", "오류"), 
            dict_tab.get("상품", '0'), 
            dict_tab.get("룩북", '0'), 
            dict_tab.get("기사", '0'),
            dict_tab.get("스타일", '0'),
            dict_tab.get("콘텐츠", '0'),
            dict_tab.get("후기", '0'),
            dict_like.get("like", '0')
           ]
    
    sql = "INSERT INTO BRAND_INFO(BRD_NAME_ENG, BRD_NAME_KOR, PRODUCT, LOOKBOOK, ARTICLE, STYLE, CONTENTS, REVIEW, LIKES) "\
         + "VALUES("\
         + "'" + dict_name["brd_name_eng"] +"'" +  ","\
         + "'" + dict_name["brd_name_kor"] +"'" +  ","\
         + "'" + dict_tab.get("상품", '0') +"'" +  ","\
         + "'" + dict_tab.get("룩북", '0') +"'" +  ","\
         + "'" + dict_tab.get("기사", '0') +"'" +  ","\
         + "'" + dict_tab.get("스타일", '0') +"'" +  ","\
         + "'" + dict_tab.get("콘텐츠", '0') +"'" +  ","\
         + "'" + dict_tab.get("후기", '0') +"'" +  ","\
         + "'" + dict_like.get("like", '0') +"'" +  ");"
    
    
#     cursor.execute(sql)    

    cursor.execute("insert into BRAND_INFO(BRD_NAME_ENG, BRD_NAME_KOR, PRODUCT, LOOKBOOK, ARTICLE, STYLE, CONTENTS, REVIEW, LIKES) values(%s, %s, %s, %s, %s, %s, %s, %s, %s)", data)
    print(sql)
    conn.commit()
```
